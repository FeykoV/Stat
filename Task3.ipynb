```python
import pandas as pd
import numpy as np
from math import factorial as fl
```

Задача 1.
Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150.
Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое,
среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
data = pd.Series([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
ME = data.sum() / data.count()
SD = np.sqrt(((data - ME) ** 2).sum() / data.count())
variance = ((data - ME) ** 2).sum() / data.count()
variance2 = ((data - ME) ** 2).sum() / (data.count() - 1)
print(f'Среднее арифмитическое = {ME :.2f}')
print(f'Среднее квадратичное отклонение = {SD :.2f}')
print(f'Смещенная дисперсия = {variance :.4f}')
print(f'Несмещенная дисперсия = {variance2 :.4f}')
```

задача 2.
В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых.
Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?


PА11 - из первого ящика вынули два белых мяча

PА12 - из второго ящика вынули один белый мяч

PА21 - из первого ящика вынули один белый мяч

PА22 - из второго ящика вынули два белых мяча

PА31 - из первого ящика вынули ноль белых мячей

PА32 - из второго ящика вынули три белых мяча

PА41 - из первого ящика вынули три белых мяча

PА42 - из второго ящика вынули ноль белых мячей


PА - из ящиков вынули три белых мяча

PА = PА11* PА12 + PА21* PА22 + PА31* PА32 + PА41* PА42


```python
PA11 = (fl(5) / (fl(2) * fl(5 - 2))) * (fl(3) / (fl(0) * fl(3 - 0))) / (fl(8) / (fl(2) * fl(8 - 2)))
PA12 = (fl(5) / (fl(1) * fl(5 - 1))) * (fl(7) / (fl(3) * fl(7 - 3))) / (fl(12) / (fl(4) * fl(12 - 4)))
PA21 = (fl(5) / (fl(1) * fl(5 - 1))) * (fl(3) / (fl(1) * fl(3 - 1))) / (fl(8) / (fl(2) * fl(8 - 2)))
PА22 = (fl(5) / (fl(2) * fl(5 - 2))) * (fl(7) / (fl(2) * fl(7 - 2))) / (fl(12) / (fl(4) * fl(12 - 4)))
PA31 = (fl(5) / (fl(0) * fl(5 - 0))) * (fl(3) / (fl(2) * fl(3 - 2))) / (fl(8) / (fl(2) * fl(8 - 2)))
PA32 = (fl(5) / (fl(3) * fl(5 - 3))) * (fl(7) / (fl(1) * fl(7 - 1))) / (fl(12) / (fl(4) * fl(12 - 4)))
PA41 = (fl(5) / (fl(3) * fl(5 - 3))) * (fl(7) / (fl(1) * fl(7 - 1))) / (fl(12) / (fl(4) * fl(12 - 4)))
PA42 = (fl(5) / (fl(0) * fl(5 - 0))) * (fl(3) / (fl(2) * fl(3 - 2))) / (fl(8) / (fl(2) * fl(8 - 2)))
```

```python
PA = PA11 * PA12 + PA21 * PА22 + PA31 * PA32 + PA41 * PA42
print(f'Вероятность того, что 3 мяча белые = {PA :.4f}')
```

Задача 3. 
На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень.
Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6.
Найти вероятность того, что выстрел произведен: 
a). первым спортсменом 
б). вторым спортсменом 
в). третьим спортсменом.

```python
PA = 1 / 3 * 0.9 + 1 / 3 * 0.8 + 1 / 3 * 0.6
PA1 = 1 / 3 * 0.9 / PA
PA2 = 1 / 3 * 0.8 / PA
PA3 = 1 / 3 * 0.6 / PA
print(f'Вероятность того, что мишень поразил первый спортсмен = {PA1 :.4f}')
print(f'Вероятность того, что мишень поразил второй спортсмен = {PA2 :.4f}')
print(f'Вероятность того, что мишень поразил третий спортсмен = {PA3 :.4f}')
```

Задача 4.
В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же,
сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8.
Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. 
Студент сдал первую сессию. Какова вероятность, что он учится: 
a). на факультете A 
б). на факультете B 
в). на факультете C?

```python
P = 1 / 4 * 0.8 + 1 / 4 * 0.7 + 2 / 4 * 0.9
PA = 1 / 4 * 0.8 / P
PB = 1 / 4 * 0.7 / P
PC = 2 / 4 * 0.9 / P
print(f'Вероятность того, что студент учится на факультете А = {PA :.4f}')
print(f'Вероятность того, что студент учится на факультете B = {PB :.4f}')
print(f'Вероятность того, что студент учится на факультете C = {PC :.4f}')
```

для второй - 0.2, для третьей - 0.25. 
Какова вероятность того, что в первый месяц выйдут из строя:
а). все детали
б). только две детали 
в). хотя бы одна деталь 
г). от одной до двух деталей?

```python
PA = 0.1 * 0.2 * 0.25
print(f'Вероятность того, что из строя выйдут все детали = {PA :.4f}')
```

```python
PB = 0.1 * 0.2 * 0.75 + 0.1 * 0.8 * 0.25 + 0.9 * 0.2 * 0.25
print(f'Вероятность того, что из строя выйдут только две детали = {PB :.4f}')
```

```python
PC = 1 - 0.9 * 0.8 * 0.75
print(f'Вероятность того, что из строя выйдет хотя бы одна деталь = {PC :.4f}')
```

```python
PD = 0.9 * 0.2 * 0.25 + 0.1 * 0.8 * 0.25 + 0.1 * 0.2 * 0.75
PE = PB + PD
print(f'Вероятность того, что из строя выйдет от одной до двух деталей = {PE :.4f}')
```
